```python
def salary_predictions():
    
    df = pd.DataFrame(index=G.nodes())
    
    #Feature Extraction
    df['Department'] = pd.Series(nx.get_node_attributes(G, 'Department'))
    df['ManagementSalary'] = pd.Series(nx.get_node_attributes(G, 'ManagementSalary'))
    df['Clustering'] = pd.Series(nx.clustering(G))
    df['Degree'] = pd.Series(G.degree())
    df['Deg_Cent'] = pd.Series(nx.degree_centrality(G))
    df['Clo_Cent'] = pd.Series(nx.closeness_centrality(G))
    df['Btn_Cent'] = pd.Series(nx.betweenness_centrality(G))
    df['PR_Scaled'] = pd.Series(nx.pagerank(G))
    df['HubScore'] = pd.Series(nx.hits(G)[0])
    df['AuthScore'] = pd.Series(nx.hits(G)[1])
    
    #Train Test Seperation
    trainDf = df[~df.ManagementSalary.isnull()]
    testDf = df[df.ManagementSalary.isnull()]
    
    #Get Dummy
    trainDf = pd.get_dummies(trainDf, columns=['Department'])
    testDf = pd.get_dummies(testDf, columns=['Department'])
    
    #Align Train/Test
    trainDf, testDf = trainDf.align(testDf, join='inner', axis=1)
    
    #Seperate X and y
    trainX = trainDf.drop(['ManagementSalary'], axis=1)
    train_y = trainDf.ManagementSalary.astype('int')
    testDf.drop(['ManagementSalary'], axis=1, inplace=True)
    
    #Train Model
    rf_clf = RandomForestClassifier(n_estimators=5000, random_state=1).fit(trainX, train_y)
    
    proba = rf_clf.predict_proba(testDf)[:, 1]
    
    submission = pd.Series(proba, index=testDf.index)
    
    return submission
```

```python
def new_connections_predictions():
    
    future_connections['PrefAttach'] = [i[2] for i in nx.preferential_attachment(G, future_connections.index)]
    future_connections['CommonNeighbors'] = future_connections.index.map(lambda node: len(list(nx.common_neighbors(G, node[0], node[1]))))
    future_connections['JacdCoef'] = [item[2] for item in nx.jaccard_coefficient(G, future_connections.index)]
    future_connections['ResAlloc'] = [item[2] for item in nx.resource_allocation_index(G, future_connections.index)]
    future_connections['Ada-AdarIdx'] = [item[2] for item in nx.adamic_adar_index(G, future_connections.index)]
    
    trainDf = future_connections[~future_connections['Future Connection'].isnull()].copy()
    testDf = future_connections[future_connections['Future Connection'].isnull()].copy()
    testDf.drop(['Future Connection'], axis=1, inplace=True)
    
    train_X = trainDf.drop(['Future Connection'], axis=1)
    train_y = trainDf.loc[:, 'Future Connection'].astype('int')
    
    logit = LogisticRegression().fit(train_X, train_y)
    
    submission = pd.Series(logit.predict_proba(testDf)[:, 1], index=testDf.index)
    
    return submission
```
