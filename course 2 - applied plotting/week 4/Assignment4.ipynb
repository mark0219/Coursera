```python
%matplotlib notebook
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df_with_children = pd.read_csv('C:\\Users\\mzhang40\\OneDrive - DXC Production\\DXC_University\\Coursera\\applied data science with Python\\course 2 - applied plotting\\week 4\\Labor force participation rate - With own children under 18.csv')
df_with_no_children = pd.read_csv('C:\\Users\\mzhang40\\OneDrive - DXC Production\\DXC_University\\Coursera\\applied data science with Python\\course 2 - applied plotting\\week 4\\Labor force participation rate - With no own children under 18.csv')
```

```python
plt.figure()
axes = plt.gca()

plt.plot(df_with_children.Year, df_with_children.Value)
plt.plot(df_with_no_children.Year, df_with_no_children.Value)
plt.title('A Comparison of U.S. Labor Participation Rate Since 1994: \nPeope With Own Children V.S. People With No Own Children')
plt.ylabel('Percentage')
plt.xticks(np.arange(1994, 2019, 2), rotation=40)
axes.set_ylim([30, 90])
axes.spines['top'].set_visible(False)
axes.spines['right'].set_visible(False)
axes.legend(['With Own Children Under 18', 'With No Own Children Under 18'], prop={'size': 10}, loc=3);
```
